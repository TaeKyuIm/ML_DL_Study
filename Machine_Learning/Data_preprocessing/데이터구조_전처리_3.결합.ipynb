# 조건에 따라 결합할 마스터 테이블 변경하기(난이도 높음, 볼사람만 볼것)
즉 특정 조건에 따라 마스터 테이블을 변경하여 하나의 테이블로 만드는 과정입니다.
예를들어 호텔 예약 시스템을 설명하자면, 어느 호텔 예약 추천사이트에서 1순위와 2순위 호텔을 추천을 해준다고 가정합시다.
전체 호텔의 개수를 n이라 하면, n x n-1이 추천 경우의 수 입니다.
그러나 이는 호텔 개수가 조금만 커도 너무 큰 처리 비용을 발생시킵니다.
따라서 호텔 데이터에서 소규모 지역단위(small_area_name)과 대규모 지역단위(big_area_name)으로 구분되는데,
작은 지역 단위로 추천을 한정해주면 비용을 줄일 수 있습니다.
그러나 작은 지역 단위에서 호텔이 없을 경우 문제가 발생하여 다른 지역의 호텔을 추천 해줄 필요가 있습니다.

```python
import gc

small_area_mst = hotel_tb.groupby(['big_area_name', 'small_area_name'], as_index=False).size()
small_area_mst
```

지금은 대규모 지역으로 묶고, 다시 소규모 지역으로 묶었습니다. 여기서 20건 미만일경우 다른 지역의 호텔도 추천을 해야하므로,
이에 활용할 데이터를 만들어보겠습니다.

```python
small_area_mst.columns = ['big_area_name', 'small_area_name', 'hotel_cnt']
small_area_mst
```

즉 여기서 20건 이상일 경우 작은 지역말고 큰 지역까지 포함해서 호텔을 추천해주어야 한다.
즉 hotel_cnt가 20보다 작을경우, big_area_name, 클 경우 small_area_name을 매칭시켜야 한다.
이를 위해선 수학적인 비교가 필요하고, numpy를 이용하면 됩니다.

```python
import numpy as np
```

```python
small_area_mst['join_area_id'] = np.where(small_area_mst['hotel_cnt'] - 1 > 20,
                                         small_area_mst['small_area_name'],
                                         small_area_mst['big_area_name'])
small_area_mst
```

where란 함수는 (조건, x,y)로 써지는데, 조건에 해당될경우 x를 연산하여 반환하고, 아닐경우 y를 연산하여 반환한다.

```python
small_area_mst.drop(['hotel_cnt', 'big_area_name'], axis=1, inplace=True)
small_area_mst
```

```python
base_hotel_mst = pd.merge(hotel_tb, small_area_mst, on='small_area_name')
base_hotel_mst
```

```python
base_hotel_mst.loc[:, ['hotel_id', 'join_area_id']]
```

따라서 거기에 걸었던 조건이 일정 지역에 호텔의 개수가 일정개수 이하면, 더 큰 지역으로 확장 한다는 것입니다.

# 과거 데이터에서 정보 얻기
주가, pos 데이터, 웹 접속 로그 등 과거의 데이터가 유용한 경우가 많습니다. 따라서 이 경우에서 과거의 데이터를 결합하여 새로운 정보를 얻는 것도 데이터 분석에 많은 도움이 됩니다.

우선 n건전의 데이터를 처리하는 것을 보겠습니다.
문제의 상황은 사용자의 과거 호텔 데이터 이용 데이터를 통해 미래의 호텔 가격을 예측하는 것에 대해 이용할 자료를 만드는 것이 목표입니다.

```python
result = reserve_tb.groupby('customer_id').apply(lambda group:
                                                group.sort_values(by = 'reserve_datetime', axis = 0, inplace = False))

result
```

groupby 메소드로 사용자별로 호텔 예약 기록을 결합하고, sort_values를 통해 시간에 대해 순서대로 정렬하였습니다.

```python
result['before_price'] = result['total_price'].groupby('customer_id').shift(periods = 2)
result
```

이제 과거의 값들을 넣을 row를 만들었으므로, 그 row에 넣을 값들을 계산하여야 합니다.
그 방법중 하나가 합계와 평균이 있는데, 우선 합계부터 구하는 것부터 계산하겠습니다.
이번에는 어느 특정 예약 날짜가 있을때, 그 예약 날짜 이전의 3건의 금액의 합을 구하는 코드를 짜볼려고 합니다.

```python
result_sum = reserve_tb.groupby('customer_id').apply(lambda x : x.sort_values(by = 'reserve_datetime', ascending = True))\
.reset_index(drop=True)

result_sum
```

우선 고객 단위로 groupby 로 묶은 다음, 예약 시간을 오름차순 순으로 묶었습니다.
그렇게 해야 특정 날짜가 정해졌을때, 그 예약 날짜 이전의 세 건의 예약 날짜와 가격을 정할 수 있습니다.

```python
result_sum['price_sum'] = pd.Series(result_sum.loc[:, ['customer_id', 'total_price']]
                                   .groupby('customer_id').rolling(center = False, window = 3, min_periods = 3).sum()
                                   .reset_index(drop = True).loc[:, 'total_price'])
result_sum

# DataFrame에서 column을 추가할때 복잡한 함수를 적용할 경우 Series를 통해 값을 넣는 것도 한번 체크를 했으면 좋겠습니다.
```

rolling 이란 함수는 데이터를 window로 나누는 함수입니다.
window로 몇 건 단위로 데이터를 나누는 지 결정을 합니다.
즉 여기의 경우 특정 예약 날짜에 대해 앞선 3건의 합을 구하기로 했으므로 window = 3 이 됩니다.
min_periods는 최소 단위가 없을 경우 연산을 진행하지 않습니다.
예를 들어 (10, 20) 과 (10, 20, 30)의 합을 구할때 각각 30, 60 인데 상당한 차이가 납니다.
min_periods함수를 쓸 경우 앞에 무조건 3건만 있는 경우에만 뒤에 있는 sum()을 계산합니다.
center = True를 이해하기 위해선 window 를 통해 데이터를 가져가는 것에 대해 우선 이해가 필요합니다.
예를들어 (10000, 20000, 30000, 12000, 32000, 40000) 이 있을때, center = True를 하면 번호를 중간부터 매깁니다.
즉 (10000, 20000, 30000)에 20000을 기준으로 0번을 매기고, 그 다음 가져갈때는 (20000, 30000, 12000)에서 30000에 1번을 매깁니다.
default 은 위의 경우에서 30000에 0번, 12000에 1번을 매깁니다.

이제 과거의 가격에서 한 건 이전 부터 세 건 이전까지 평균 예약 금액을 계산하고, 세 건보다 밑일경우 존재하는 건수에 대해서 존재하는 건수에 대해서 계산하는 경우를 한번 살펴보겠습니다.

```python
result_avg = reserve_tb.groupby('customer_id').apply(lambda x : x.sort_values(by = 'reserve_datetime', ascending =True))\
.reset_index(drop =True)

result_avg
```

```python
result_avg['price_avg'] = pd.Series(result_avg.groupby('customer_id')
                                   ['total_price'].rolling(center = False, window = 3, min_periods = 1).mean()\
                                   .reset_index(drop = True))

result_avg
```

여기서 평균값 계산은 다 끝났으나, groupby 데이터프레임에서 처음으로 호텔에 갔을때는 값이 없어야하므로
shift()함수를 통해 1 row 식 내려주면 우리가 생각한데로 자료를 구성할 수 있습니다.

```python
result_avg['price_avg'] = result_avg.groupby('customer_id')['price_avg'].shift(periods = 1)
result_avg
```

즉 제일 위에 NaN값이 나오므로 제대로 한 것이라고 생각할 수 있습니다.

앞에 경우들은 계산을 할 건수가 명확하지만, 이제는 건수가 명확하지 않은 경우에 대해 합을 구하는 것을 해보겠습니다.
예를들어 호텔자료에서 특정 시간동안 누구는 3번, 누구는 1번, 누구는 5번 올 수 있는데, 이에 대해 합계값을 구하는 것을 해보겠습니다.
한번도 안왔으면 0을 반환합니다.

```python
import pandas.tseries.offsets as offsets
import operator

reserve_tb['reserve_datetime'] = pd.to_datetime(reserve_tb['reserve_datetime'], format= '%Y-%m-%d %H:%M:%S')

reserve_tb
```

```python
sum_table = pd.merge(
reserve_tb[['reserve_id', 'customer_id', 'reserve_datetime']],
reserve_tb[['customer_id', 'reserve_datetime', 'total_price']].rename(columns={'reserve_datetime' : 'reserve_datetime_before'}),
on = 'customer_id')
sum_table
```

```python
sum_table = sum_table[operator.and_(
sum_table['reserve_datetime'] > sum_table['reserve_datetime_before'],
sum_table['reserve_datetime'] + offsets.Day(-90) <= sum_table['reserve_datetime_before'])].groupby('reserve_id')['total_price']\
.sum().reset_index()

sum_table
```

```python
sum_table.columns = ['reserve_id', 'total_price_sum']
pd.merge(reserve_tb, sum_table, on='reserve_id', how = 'left').fillna(0)

sum_table
```

# 상호결합
앞선 결합의 경우 결합 키가 있는 경우를 소개했는데, 몇몇의 경우에서 결합 키를 지정하지 않는 전체 결합이 필요한 경우가 있습니다.
상호결합은 결합에 사용되는 테이블을 모두 생성하여 하는 결합입니다. 보통 집계나 학습 데이터를 만들기 위한 전처리에 사용됩니다.

```python
import datetime
from dateutil.relativedelta import relativedelta

month_mst = pd.DataFrame({'year_month' : [(datetime.date(2017, 1, 1) + relativedelta(months = x)).strftime("%Y%m")
                                         for x in range(0, 3)]})

month_mst
```

relativedelta 라는 함수로 날짜의 상대적인 계산이 가능합니다.
즉 2017년 1월 ,2월, 3월을 테이블로 만들었습니다.

```python
customer_tb = pd.read_csv('c:/Users/User/daejeon/customer.csv')
customer_tb
```

```python
customer_tb['join_key'] = 0
month_mst['join_key'] = 0
```

상호결합 하기 위해 임의적인 결합용 키 column을 만들었습니다.

```python
customer_mst = pd.merge(customer_tb[['customer_id', 'join_key']], month_mst, on='join_key')
customer_mst
```

```python
reserve_tb['year_month'] = reserve_tb['checkin_date'].apply(lambda x: pd.to_datetime(x, format = '%Y-%m-%d').strftime('%Y%m'))

reserve_tb
```

to_datetime()이란 함수를 사용하여 문자열로 된 날짜를 timestamp로 바꾸었습니다. 그 다음 strftime()함수를 통해 연도-월 형태로 범주화 했습니다.

```python
summary_result = pd.merge(customer_mst, reserve_tb[['customer_id', 'year_month', 'total_price']],
                         on = ['customer_id', 'year_month'], how = 'left').groupby(['customer_id', 'year_month'])['total_price']\
.sum().reset_index()

summary_result
```

묶을때 연도월과 고객단위로 묶어서 left 결합을 통해 만약에 해당 월에 호텔을 안갔으면 NaN이 뜨도록 했습니다.

```python
summary_result.fillna(0, inplace = True)
summary_result
```
