```python
x[0, :]
# 첫번째 행을 슬라이싱
```

# Join(cat, stack) 기능 사용하기

```python
x = torch.cuda.FloatTensor([[1, 2, 3], [4, 5, 6]])
y = torch.cuda.FloatTensor([[-1, -2, -3], [-4, -5, -6]])
z1 = torch.cat([x, y], dim=0)
z1
```

```python
z2 = torch.cat([x, y], dim=1)
z2
# 즉 dim으로 가로로 합칠지 세로로 합칠지 결정
```

```python
x_stack = torch.stack([x, x, x, x], dim=0)
print(x_stack.size())
x_stack
```

```python
x_stack_1 = torch.stack([x, x, x, x], dim = 1)
print(x_stack_1.size())
x_stack_1
```

# slicing 기능 사용하기

```python
x_1, x_2 = torch.chunk(z1, 2, dim=0)
y_1, y_2, y_3 = torch.chunk(z1, 3, dim =1)
# tensor를 몇부분으로 나누는 함수
```

```python
print(x_1)
print(x_2)
```

```python
print(y_1)
print(y_2)
print(y_3)
```

# squeezing 기능 사용하기

```python
x = torch.zeros(2, 1, 2, 1, 2)
x.size()
```

```python
y = torch.squeeze(x)
y.size()
# 치수가 1인 차원이 다 날아감
```

```python
y = torch.squeeze(x, 0)
y.size()
```

```python
y = torch.squeeze(x, 1)
y.size()
# 즉 두번째 매개변수는 지정된 차원만 압축한다.
```

```python
x = torch.zeros(2, 3, 4)
torch.unsqueeze(x, 0).size()
```

```python
x = torch.tensor([1, 2, 3, 4])
torch.unsqueeze(x, 0)
```

```python
torch.unsqueeze(x, 1)
```

```python
torch.unsqueeze(x, 1).size()

# 즉 unsqueeze로 차원을 치수가 1인 차원을 추가 가능
```

# Initialization

```python
import torch.nn.init as init
```
