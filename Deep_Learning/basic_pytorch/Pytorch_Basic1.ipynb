# 텐서의 차원 조작

```python
x1.view(4, 1)
# 2x2를 4x1 으로
```

```python
x1.reshape(4, 1)
```

```python
x1.view(-1)
# 2x2을 1차원 벡터로
```

```python
x1.reshape(-1)
# tensor([1,2,3,4])
```

```python
x2.t()
# 전치행렬
```

```python
torch.cat([x1, x2], dim=1)
```

# Numpy to Tensor 또는 Tensor to Numpy

```python
import numpy as np
```
